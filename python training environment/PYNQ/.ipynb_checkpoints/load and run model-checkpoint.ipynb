Load a trained LSTM and the DROPBEAR dataset, the perform a forward pass. The model I'm using is a 40-unit, 1-cell model which sampled DROPBEAR at 500 us.

TensorFlow 2.5.0

Numpy 1.19.5

```python
import tensorflow.keras as keras
import matplotlib.pyplot as plt
import numpy as np
import time
model = keras.models.load_model("./model")
X = np.load("X.npy")
y = np.load("y.npy")
t = np.linspace(0, y.size*.0005, y.size, endpoint=False)
```

Do forward pass

```python
start_time = time.perf_counter()
y_pred = model.predict(X)
stop_time = time.perf_counter()
inference_time = stop_time - start_time
print('inference time: %.4f s'%inference_time)
```

```python
plt.figure()
plt.plot(t, y.flatten(), label='reference')
plt.plot(t, y_pred.flatten(), alpha=.8, label='prediction')
plt.xlim((t[0], t[-1]))
plt.legend()
plt.xlabel('time (s)')
plt.ylabel('displacement (normalized)')
plt.tight_layout()
```
